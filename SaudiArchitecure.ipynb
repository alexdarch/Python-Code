Things to note about the data:
* Guest rooms, land size, living room and building age have small but significant outliers, far larger/more frequent than would be expected from any standard distribution. Most of the distributions have **heavy tails** - defined as a tail probability decaying less quickly than an exponential distribution - i.e. they have high outlier probabilities. This may cause problems for machine learning algorithms, however, it is often easy to transform variables like this to get more bell-shaped distributions.
* Location is discrete (and doesnt have any resembelence to where it actually is). Therefore a one-hot encoding should be used.
* Living room is the "dodgy data" mentioned earlier. The survey meant the hall, but most people thought it was the lounge
* There are no nan/empty data columns, unless you count living room. The solution to this could just be to remove the values at 356. Alternatively as > 90% of people thought this was the lounge, the data could be relabelled (however, this is pointless as there is only one value!)
* Building age and land size both have very large values. It would be sensible to normalise these to a more managable scale for a machine learning algorithm.

The first thing to do for any machine learning task is to split the data into a training and a test set. Since this dataset is quite small (868 entries) there is a significant amount of sampling bias (e.g. only 1.7% of values in catagory 8, yet there are 5% in reality), therefore we could use stratified sampling. Generally the most *important* attribute would be used for stratifying, however, since I'm not sure what that is, I will use age as age is both the closest to a uniform distribution and a continuous distribution (i.e. it has a good spread over the data).

```python
arch['age_category'] = pd.cut(arch['age'], 
                              bins = [0.0, 30.0, 40.0, 50.0, 60.0, 70.0, 80.0, np.inf], # min=20.0, max = 89.0
                              labels = [2, 3, 4, 5, 6, 7, 8])
arch['age_category'].hist(bins=6, density=True)
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(arch, test_size=0.2, random_state=42)  # seed=42 so test set doesnt change each run
print(len(train_set), len(test_set))
test_set['age_category'].value_counts(sort=False) / len(test_set)
```

```python
# stratified sampling
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(arch, arch['age_category']):
    strat_train_set = arch.iloc[train_index]
    strat_test_set = arch.iloc[test_index]
```

```python
strat_test_set['age_category'].value_counts(sort=False) / len(strat_test_set)  # much better!
```

Finally, drop the age_category column

```python
strat_train_set = strat_train_set.drop('age_category', axis=1)
strat_test_set = strat_test_set.drop('age_category', axis=1)
```

## 3. Data Preparation  <a name="data_preparation"> </a>

Data preparation usually involves 3 steps: 

1. Clean the data from null/inf values, or any biases
2. Handle text and categorical data
3. Rescale any unweildy features

Firstly, we need to separate the training set into the input variables/features and the target variables/features (or labels).

```python
train_inputs = strat_train_set.drop(['male_guest_room', 'female_guest_room', 'living_room'], axis=1)
train_targets = strat_train_set[['male_guest_room', 'female_guest_room', 'living_room']].copy()
```

### 3.1 Cleaning the data

The data supplied has a number of quirks, as previously shown. These include the rounding of the number of days that each room has been used, and the ambiguity of the `living_room` responses (the responses were split between people who assumed that the questions meant the lounge, and the entrance hall - which was the intended meaning - hence there is a massive skew to 365 days if the lounge interpretation was made). There are a number of possible ways to deal with this:

For the `living_room` we could:
* Remove the entire attribute.
* Split the data in two, and have a number of unfilled values for each of those catagories. However, having a `lounge` attribute will give no interesting data as all answers will be 365. This leaves a partially filled `living_room` attribute.
* With a partially filled living room attribute we could remove the empty rows, set the empty values to zero or to the median value (use `from sklearn.impute import SimpleImputer`).

For the rounded attributes we could:
* Add some noise to each datapoint, e.g. $x_{new} = x_{old} + \mathbb{U}(-5, 5)$, a uniform random number in the range $-5 < X < 5$.
* Round the data using the .cut() method in pandas, and therefore treat it as a catagorical feature.
* Leave the data as-is.

However, as I am not sure what to do with these now, they have just been left as is.

### 3.2 Categorical Locations

If the location attribute was given as text we could us sklearn's `OrdinalEncoder` class, however, this assigns each catagory a number and this does not make sense for the location category as this is an **unordered category** (an ordered category would be "bad", "average", "good", "excellent", where each category is "close" to its neighbours). The standard way to deal with unordered categories is to use a one-hot encoding. We can use sklearn's `OneHotEncoder` class to easily do this.

Note `OneHotEncoder`'s `fit_transform` method creates a SciPy *sparse martix*, which stores the value and location of each entry rather than the full matrix (which can be very memory intensive).

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder(categories='auto')  # create an instance of the OneHotEncoder class
locations = [[i] for i in train_inputs['location'].values]  # need values to be a list of lists
location_one_hot = cat_encoder.fit_transform(locations)
print(location_one_hot.toarray())
```

```python
cat_encoder.categories_
```

### 3.3 Feature Scaling

Input variables of ML algorithms almost always need to be scaled or transformed in order to make them useable. Here we are aiming to get values within (or near enough) the -1 to +1 range, and also we aim to transform each feature such that it either looks like a uniform or a normal distribution. The target features do not usually need to be changed.

The two most common ways of scaling features are min-max scaling and standardisation (sometimes called normalisation). Standardisation is simply subtracting the mean of the data and dividing by the standard deviation, and is therefore much less affected by outliers than min-max scaling. Since this dataset has many outliers, we shall use standardisation. Note that it is important to only fit these transformations to the training data and not the test data.

```python
from sklearn.preprocessing import StandardScaler
```

### 3.4 Pipelining

A nice way of connecting all of these things together is to use sklearn's pipelining functionality. Pipelining allows us to sequence transformations in a way such that reusing and editing these transformations is simple. Additionally sklearn provides a class to interact with pandas easily called ColumnTransformer.

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

We need to define a pipeline for the numerical attributes and the categorical attributes seperately, and then combine them.

```python
num_steps = [
    # ('data_cleaner', DataCleaner()),   # some list of Estimators and Transformers
    ('std_scaler', StandardScaler()),  # all but the last one must have a .fit_transform() method
]
num_pipeline = Pipeline(num_steps)
```

```python
attributes = train_inputs.columns

num_attributes = list(train_inputs.drop(['location', 'timestamp'], axis=1).columns)
cat_attributes = ['location']
print(num_attributes, cat_attributes)

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attributes),
    ('cat', OneHotEncoder(categories='auto'), cat_attributes),
])
```

```python
train_inputs_prepared = full_pipeline.fit_transform(train_inputs)
train_inputs_prepared.shape
```

```python
prepared = pd.DataFrame(train_inputs_prepared, columns=['family_member', 'age', 'male_members',
       'female_members', 'building_age', 'land_size', 'outside_room', 'total_family_members', 'l1', 'l2', 'l3', 'l4', 'l5'])
prepared.head()
```

### 4.1 Basic Models

```python
lin_reg = LinearRegression()
lin_reg.fit(train_inputs_prepared, train_targets)
```

Fitting a model can be as simple as that. Now all that is needed is to see how it did. We can do this by just doing some predictions on a subset of the training data, or by looking at a performance metric, such as RMSE (root mean squared error).

This is tiny! Clearly this is overfitting.

### 4.2 Cross-Validation and Ensemble Methods

Cross validation splits the dataset into k *folds*, or subsets. The model is then trained k times, picking a different fold as the test set each time. This allows for better evaluation of the model in general. Note that sklearn's cross-validation requires a utility function rather than a cost function so we just negate the mean squared error.

```python
tree_scores = cross_val_score(tree_reg, train_inputs_prepared, train_targets, scoring='neg_mean_squared_error', cv=10)
tree_rmses = np.sqrt(-tree_scores)
lin_scores = cross_val_score(lin_reg, train_inputs_prepared, train_targets, scoring='neg_mean_squared_error', cv=10)
lin_rmses = np.sqrt(-lin_scores)

print("Tree: ", tree_rmses, tree_rmses.mean(), tree_rmses.std())
print("Lin: ", lin_rmses, lin_rmses.mean(), lin_rmses.std())
```

The tree regressor is now even worse than the linear regressor! Note that cross-validation also allows us to measure how precise the estimate of the error is, with the standard deviation.

Often we can get very good performance using one of these models that are poor on their own by using an **Ensemble Method**. Random forests are an ensemble method for decision trees that work by training many decision trees on random subsets of data and averageing out their predictions.

```python
forest_reg = RandomForestRegressor()
forest_reg.fit(train_inputs_prepared, train_targets)

forest_scores = cross_val_score(forest_reg, train_inputs_prepared, train_targets, scoring='neg_mean_squared_error', cv=10)
forest_rmses = np.sqrt(-forest_scores)
print("Forest: ", forest_rmses, forest_rmses.mean(), forest_rmses.std())
```

This actually performed even worse than the regular decision tree regressor. However, the result is more general, which suggests that these models are not complex enough for the data at hand - or that the input data does not capture any relationship with the target data.

### 4.3 Hyperparameter Tuning

Hyperparameters are paramters govern how the model is set up, e.g. the use of 10 folds in the cross_val above, or the squared exponent in the linear regressor's discriminator - they can be anything that you set before learning begins. Tuning them is a good way of improving your model.

```python
param_grid = [
    {'n_estimators': [3, 10, 30, 100], 'max_features': [2, 4, 6, 8]},  # try to go up in powers of 10 (note log(3) ~ 0.5 so these increase as 0.5, 1, 1.5, ... in the log scale)
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},    
]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                          scoring='neg_mean_squared_error',
                          return_train_score=True)
grid_search.fit(train_inputs_prepared, train_targets)
```

```python
print(grid_search.best_params_)  # sklearn stores its learnt parameters in attributes with an underscore suffix
print(grid_search.best_estimator_)
res = grid_search.cv_results_
for mean_score, params in zip(res['mean_test_score'], res['params']):
    print(np.sqrt(-mean_score), params)
```

Since 300 is the max n_estimators tries, it is probably a good idea to gridsearch through higher values. Still not much improvement, but this was expected as I don't think the data is sufficient. However, this does show you how to conduct a grid search. Another (better if lots of hyperparams) way of doing this is through a randomised search `RandomizedSearchCV`.

### 4.4 Analyse the Model

```python
feature_importances = grid_search.best_estimator_.feature_importances_
#print(feature_importances)
cat_encoder = full_pipeline.named_transformers_['cat']
cat_one_hot_attributes = list(cat_encoder.categories_[0])
attributes = num_attributes + cat_one_hot_attributes

sorted(zip(feature_importances, attributes), reverse=True)
```

From this we can see that `building_age`, `age` and `land_size` are the most important catagories. The location category is of almost no importance. However, take this with a grain of salt, as the model has not performed better than by just taking the mean of the target values.

## 5. Evaluate <a name='evaluate'></a>

Finally we evaluate the model on the test set. To do this we call `.transform()`, not `.fit_transform()`.

```python
final_model = grid_search.best_estimator_
test_inputs = strat_test_set.drop(['male_guest_room', 'female_guest_room', 'living_room'], axis=1)
test_targets = strat_test_set[['male_guest_room', 'female_guest_room', 'living_room']].copy()

test_prepared = full_pipeline.transform(test_inputs)
final_predictions = final_model.predict(test_prepared)
```

```python
final_mse = mean_squared_error(test_targets, final_predictions)
final_rmse = np.sqrt(final_mse)
print(final_rmse)
```

The performance is expected to be slightly worse than in the cross_validation phase. In this instance, it is slightly better which suggests that the model is not very good at all.
